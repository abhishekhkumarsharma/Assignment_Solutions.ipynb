## Question 1 : What is Information Gain, and how is it used in Decision Trees?

- Information Gain (IG) measures how much uncertainty (entropy) is reduced after splitting the data on a feature.  
The feature with the highest IG is chosen for the split.  
It helps the tree decide the best feature at each step.


## Question 2: What is the difference between Gini Impurity and Entropy?

- **Entropy** uses logarithms and measures disorder.  
- **Gini Impurity** measures misclassification probability and is faster.  
- Gini is used in CART; Entropy in ID3/C4.5.


## Question 3: What is Pre-Pruning in Decision Trees?

- Pre-pruning stops tree growth early using constraints like max_depth, min_samples_split, etc., preventing overfitting.


**Question 4:Write a Python program to train a Decision Tree Classifier using Gini**
**Impurity as the criterion and print the feature importances (practical).**

```python

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

data = load_iris()
X, y = data.data, data.target

clf = DecisionTreeClassifier(criterion='gini')
clf.fit(X, y)

pd.DataFrame({'Feature': data.feature_names,
              'Importance': clf.feature_importances_})
```

**Question 7:  Write a Python program to train two SVM classifiers with Linear and RBF**
**kernels on the Wine dataset, then compare their accuracies. **
**Hint:Use SVC(kernel='linear') and SVC(kernel='rbf'), then compare accuracy scores after fitting**
**on the same dataset.

```python

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = load_wine()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

linear = SVC(kernel='linear')
rbf = SVC(kernel='rbf')

linear.fit(X_train, y_train)
rbf.fit(X_train, y_train)

acc_linear = accuracy_score(y_test, linear.predict(X_test))
acc_rbf = accuracy_score(y_test, rbf.predict(X_test))

acc_linear, acc_rbf
```

**Question 10:  Breast Cancer Dataset**
**Write a Python program to train a Gaussian Na誰ve Bayes classifier on the Breast Cancer**
**dataset and evaluate accuracy.**
**Hint:Use GaussianNB() from sklearn.naive_bayes and the Breast Cancer dataset from**
**sklearn.datasets.**

```python

from sklearn.datasets import load_breast_cancer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

gnb = GaussianNB()
gnb.fit(X_train, y_train)

accuracy_score(y_test, gnb.predict(X_test))
```
